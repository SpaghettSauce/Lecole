## Библиотека 

Вот мы подошли к использованию лучшего изобретения в мире программирования, **библиотеки**
Библиотеки это, грубо говоря, сбоник до этого написанного кода, в котором написаны какие либо функции, классы и тд.

1. Зачем нам нужны библиотеки ?

    Библиотеки используются для того чтобы пользоваться уже готовыми функциями и классами значительно упрощая работу и количество кода. 
    Так, если мне в программе надо высчитать квадратный корень то мне надо знать что такое метод Ньютона и написать данный код: 


```python
def sqrt(x):
    if x < 0:
        raise ValueError("Квадратный корень из отрицательного числа не определён")
    if x == 0:
        return 0

    guess = x / 2.0
    tolerance = 1e-10 

    while True:
        better_guess = (guess + x / guess) / 2
        if abs(guess - better_guess) < tolerance:
            return better_guess
        guess = better_guess

print(sqrt(25))   
```
