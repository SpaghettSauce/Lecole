# 1. Списки в Python
Список (list) — это упорядоченные элементы. Можно хранить числа, строки и любые объекты.

В списки, как и в переменных, данные можно приписывать по разному

```python
a = [5, 2, 10, 1]
b = []               # пустой список
n = int(input())
arr = [int(input()) for _ in range(n)]

print(a)
print(b)
print(arr)
```

Сама особенность списков заключается в том, что у каждого элемента есть своё назначенное место, от 0 до размера списка.

Так в списке a = [5,2,10,1], 5 имеет место 0, 2 имеет место 1 и тд. Само "место", прописывается в квадратных скобках

```python
a = [5, 2, 10, 1]
print (a[0])  # вывод первого элемента списка(5)
print (a[2])  # вывод третьего элемента списка(10)
```

Эта особенность списков, создает небольшую особенность как мы взаимодействуем с элементами внутри него, если мы хотим создать список, и ввести внутрь него цифры, мы не можем просто написать:

```python
a[5] = int(input())
```

Как видно, выходит ошибка, вместо этого нам надо использовать штуку которую мы проходили до этого, in range:

```python
a = [int(input()) for _ in range(5)] # создание списка с вводом 5 элементов
print (a)
```

Для вывода списка на экран, in range использовать не надо, просто пишем print(a)

# 2. Действия с списками
У списков есть несколько различных операций которые мы можем с ними выполнить, а именно:
1. Длина списка: len

```python
a = [5, 2, 10, 1]
print (len(a))  # длина списка
```

2. Добавить элемент к уже существующему списку: append

```python
a = [5, 2, 10, 1]
a.append(7) # добавляем элемент 7 в конец списка
print(a)
```

3. Сортировать список по возростанию: sort


```python
a = [5, 2, 10, 1]
a.sort()  # сортируем список по возрастанию
print(a)
```

4. Сортировка по убыванию

```python
a = [5, 2, 10, 1]
a.sort(reverse=True)  # сортируем список по убыванию
print(a)    
```
